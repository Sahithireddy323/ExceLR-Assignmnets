```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("C:\\Users\\nimma\\OneDrive\\Desktop\\Cardiotocographic.csv")
df
```

```python
df_clean = df.dropna() #Cleaning the data
df_clean
```

```python
df_clean.info()
```

```python
df_clean.describe() #mean,median
```

```python
iqr = df_clean.quantile(0.75) - df_clean.quantile(0.25) #IQR
print("Interquartile Range (IQR):")
print(iqr)
```

```python
def remove_outliers(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    return df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]

df_no_outliers = remove_outliers(df_clean)
print("Shape after removing outliers:", df_no_outliers.shape)
```

```python
df_no_outliers.hist(bins=15) #Histogram
plt.tight_layout()
plt.show()
```

```python
for col in df_no_outliers.select_dtypes(include=np.number).columns: #Boxplot
    sns.boxplot(x=df_no_outliers[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
plt.figure(figsize=(10, 8))    #Correlation and relationships
sns.heatmap(df_no_outliers.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.scatterplot(data=df_no_outliers, x='UC', y='LB') #scatter plot
plt.title('Scatter Plot: UC vs LB')
plt.show()
```

```python
sns.violinplot(data=df_no_outliers[['ASTV', 'ALTV']]) #Violin  plot
plt.title('Violin Plot of ASTV and ALTV')
plt.show()
```
